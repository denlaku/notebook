# 排序
+ 按标签
+ 按实际值

```python
import pandas as pd
import numpy as np
```

```python
arr = [
    [1, 30, 0],
     [11, 5, 4],
     [9, 13, 3],
     [12, 8, 2],
     [10, 2, 7],
]
df = pd.DataFrame(arr, columns=['a', 'c', 'b'], index=['f', 'g', 'c', 'a', 'b'])
df
```

```python
df.sort_index(axis=1)
```

```python
df.sort_index(axis=0)
```

```python
df.sort_values(by=['a'])
help(df.sort_values)
```
