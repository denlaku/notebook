```python
import pandas as pd
import numpy as np

df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'name': ['A', 'B', 'C'],
    'age': [11,22,13]
})

df1 = pd.concat([df1, df1])

from sklearn.preprocessing import PolynomialFeatures
poly_features = df1[['age']]
poly_transformer = PolynomialFeatures(degree = 3)
poly_transformer.fit(poly_features)
poly_features = poly_transformer.transform(poly_features)

poly_features = pd.DataFrame(poly_features, columns = poly_transformer.get_feature_names(['age']))

ID = np.array(df1['ID'])
poly_features['ID'] = ID
df1
poly_features

df = df1.merge(poly_features, on='ID', how='left')
df

a = df[df['name']=='B']
sub = pd.DataFrame()
sub['name']=a['name'].reset_index(drop=True)

age_df = pd.DataFrame({
    'age': [11,22,13]
})

sub['age'] = age_df

sub
```
