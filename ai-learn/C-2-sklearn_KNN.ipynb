```python
# -*- coding: utf-8 -*-
"""
Created on Tue Jan 15 22:01:53 2019

@author: hacke
"""

import pandas as pd

data = pd.read_csv('vehicle.csv')

labels = data['label']
data = data.drop(['label'],axis = 1)

from sklearn.model_selection import train_test_split
feature_train, feature_test, label_train, label_test = train_test_split(data, labels,test_size = 0.2)


from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=9)
model.fit(data, labels) 

prediction = model.predict(feature_test)

sub = pd.DataFrame()
sub['test'] = label_test
sub['predict'] = prediction
sub['t-f'] = prediction == label_test
sub
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import classification_report
labels = ['car', 'truck']
classes = ['car', 'truck']
r = classification_report(label_test, prediction, target_names=classes, labels=labels, digits=6)
print(r)

from sklearn.metrics import precision_score
precision_score(prediction, label_test,average='weighted')
# ?classification_report
```
